### A partir d'aqui comencem a crear columnes pel nostre dataset

```python
# Create a function for the values of our new column
def quiniela(resultat):
    if resultat["GolesLocal"] > resultat["GolesVisitante"]: 
        return "1" 
    elif resultat["GolesLocal"] < resultat["GolesVisitante"]: 
        return "2"
    else: 
        return "X"

# Create a new column with the result of the match in "quiniela" style
resultados["Quiniela"] =  resultados.apply(quiniela, axis = 1)

# Check is done correctly
resultados.head(10)
```

```python
# Create a function for the values of our new column
def puntosL(resultat):
    if resultat["GolesLocal"] > resultat["GolesVisitante"]: 
        return 3 
    elif resultat["GolesLocal"] < resultat["GolesVisitante"]: 
        return 0
    else: 
        return 1

# Create a function for the values of our new column
def puntosV(resultat):
    if resultat["GolesLocal"] > resultat["GolesVisitante"]: 
        return 0 
    elif resultat["GolesLocal"] < resultat["GolesVisitante"]: 
        return 3
    else: 
        return 1

# Create new columns of how many points get Local team and another with the away team
resultados["LP"] =  resultados.apply(puntosL, axis = 1)
resultados["VP"] =  resultados.apply(puntosV, axis = 1)

#Show first rows
resultados
```

```python
# Create new variable with the names of the teams
teams = resultados.Local.unique()

# Create a bucle to itirate for each local team assign the mean of goals done in last 10 matches as a local team before this one. 
for team in teams:
    resultados.loc[resultados['Local'] == team, 'L10gf_cL'] = resultados[resultados['Local'] == team]['GolesLocal'].rolling(10).mean().shift(1)
```

```python
# Import date library
import datetime

# Transforme dtype and format of date column
resultados["Fecha"] = pd.to_datetime(resultados["Fecha"], format = "%d/%m/%Y")

# Create a function to assign each date to a period of the season (decided by my criteria that could influence into results)
def periode(Fecha):
    if Fecha.month >= 8 and Fecha.month <= 10:
        return "1"
    elif Fecha.month >= 11 or Fecha.month <= 1:
        return "2"
    elif Fecha.month >= 2 and Fecha.month <= 4:
        return "3"
    else:
        return "4"

# Create the column with the function applied    
resultados["Periode"] = resultados["Fecha"].apply(periode)

# Check it works/did it correctly
resultados[resultados["Periode"] == "2"][["Temporada", "Fecha", "Periode"]].head(25)
```

```python
# Import csv file and create a new df
classificacio = pd.read_csv("clasificacio.csv", sep=";", encoding="latin-1")

# Show it
classificacio
```

```python
# Show how many seasons are
print("Hi ha", classificacio["Temporada"].nunique(), "temporades.")
```

```python
# Create a new variable with all the teams
c_teams = classificacio["Equipo"].unique()

# Create a new bucle to know the position mean of last 10 seasons before the actual and assign in a new column.
for team in c_teams:
    classificacio.loc[classificacio['Equipo'] == team, '10_Clas'] = classificacio[classificacio['Equipo'] == team]['Posicion'].rolling(10).mean().shift(1)

# Check it is working correctly 
classificacio[classificacio["Equipo"] == "Barcelona"].head(15)
```

```python
# Print max and min of the new column to know the range
print(classificacio["10_Clas"].max())
print(classificacio["10_Clas"].min())
```

```python
# Create a function a apply a handicap depending of the position of each time in last 10 season.
def handicap(x):
    if x <= 3:
        return "1"
    elif 3 < x <= 7:
        return "2"
    elif 7 < x < 14:
        return "3"
    else:
        return "4"

# Create the new column applying the handicap function
classificacio["Handicap"] = classificacio["10_Clas"].apply(handicap)

# Check it is working correctly
classificacio[classificacio["Equipo"] == "Madrid"]
```
